```python
import pandas as pd
```

```python
df = pd.read_csv("customer_shopping_behavior.csv")
```

```python
# replace missing values with median of that particular category of item
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
# converting column names to snake casing for easier referencing
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
# create a column age_group
labels = ['young adult','adult','middle aged','senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# creating column purchase_frequency_days

frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Annually':365,
    'Every 3 Months':90
}
df['purrchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# check redundant features
(df['discount_applied']==df['promo_code_used']).all()
```

```python
# drop redundant feature
df = df.drop('promo_code_used',axis=1)
```
